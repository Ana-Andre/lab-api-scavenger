```python
'''
### Challenge 3: Hidden Cold Joke

Using Python, call Github API to find out the cold joke contained in the 24 secret files in the following repo:

https://github.com/ironhack-datalabs/scavenger

The filenames of the secret files contain `.scavengerhunt` and they are scattered in different directories
of this repo. The secret files are named from `.0001.scavengerhunt` to `.0024.scavengerhunt`. 
They are scattered randomly throughout this repo. You need to **search for these files by calling the 
Github API**, not searching the local files on your computer.
'''
```

```python
#import libraries
import json
import requests
import numpy as np
import pandas as pd
from pandas.io.json import json_normalize
```

```python
# api to dataframe
url3 = 'https://api.github.com/repos/ironhack-datalabs/scavenger/contents/'

response = requests.get(url3)
#print (response)
norm_json = json_normalize(response.json())
df_scav = pd.DataFrame(norm_json)
df_scav.head()
```
